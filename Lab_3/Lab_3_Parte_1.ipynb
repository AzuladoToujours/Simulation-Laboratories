1.2 Teniendo en cuenta el tipo de problema y el número de muestras por clase, ¿qué metodología de validación debería usar?¿Porqué?:

**R//:** K-folds estratificado ya que el sistema se encuentra desbalanceado y queremos evitar que una clase esté sobrerepresentada al momento del entranamiento y ocasione sesgo a nuestro modelo.


**Definición usada en el lab 2:**

El método de validación empleado en este caso fue K-folds estratificado ya que reorganizamos los datos de manera que cada fold tenga una buena representación de todo el conjunto de datos. lo que obliga a que cada fold tenga al menos m instancias de cada clase. Esto nos garantiza que una clase de datos no esté sobrerrepresentada, especialmente cuando el sistema esta desbalanceado. Es importante destacar que la validación cruzada estratificada ayuda a reducir el sesgo y la varianza.

En la siguiente celda defina una función que tome como entradas una matriz X y una matriz Y, entrene un modelo GMM por cada clase y retorne el listado de modelos para cada clase. Adicionalmente construya una función que use los modelos entrenados en la función anterior para hacer la clasificación de un conjunto nuevo de muestras.

```python
from sklearn.mixture import GaussianMixture

def GMMClassifierTrain(X,Y,M,tipo):    
    #X: Matriz con las características
    #Y: Matriz con las clases
    #M: Número de componentes
    #tipo: Tipo de matriz de covarianza
    nClases = np.unique(Y)
    GMMs = [] # Vector que tiene los modelos GMM entrendao
    for c in nClases:
        gmm = GaussianMixture(M,tipo)#Configurar el modelo
        Xc = X[Y==c] # Seleccionamos la matrix de la clase a entrenar
        GMMs.append(gmm.fit(Xc))        
    return GMMs #Debe retornar un objeto que contenga todos los modelos entrenados

def GMMClassfierVal(GMMs,Xtest):
    nGmm = len(GMMs) #Numero de modelos que existen
    nSamples = len(Xtest) #Numero de muestras a clasificar
    """Creamos una matriz de tamaño nSamples,nGmm en donde
    cada valor de la columna representa la probabilidad de que esta muestra pertenezca a dicha clase
    la pocision del mayor valor de la columna es la clase a la que se predice pertenecera
    """
    mProbabilities = np.zeros((nSamples,nGmm))
    for i in range(nGmm):
        mProbabilities[:,i]=GMMs[i].score_samples(Xtest)
    Yest = np.argmax(mProbabilities,axis=1)    
    return Yest #Debe retornar un vector con las clases predichas para cada una de las muestras en Xtest, de acuerdo con los modelos almacenados en GMMs
```

```python
from sklearn.datasets import load_iris
data = load_iris()
X=data.data
Xclustering = X[:,1:3]
plt.scatter(Xclustering[:,0],Xclustering[:,1])
plt.title("Datos para agrupar")
plt.show
```

```python
def implementationGMMClustering(Xclustering,M,tipo):
    gmm = GaussianMixture(M,tipo)
    gmm.fit(Xclustering)
    Ypredit = gmm.predict(Xclustering)
    plt.figure()
    titulo ='Matrix de tipo', tipo, 'con ',str(M),'Clusters'
    plt.title(titulo)
    plt.xlabel('Caracteristica 1')
    plt.ylabel('Caracteristica 2')
    plt.scatter(Xclustering[:,0],Xclustering[:,1], c=Ypredit)    
```

```python
mC = (2,2,3,3)
mType = ('full','diag','full','diag')
for m,tipo in zip(mC,mType):
    implementationGMMClustering(Xclustering,m,tipo)
```

```python
from sklearn.cluster import KMeans
def implentationKmeans(Xclustering,n_clusters):
    kmeansModel = KMeans(n_clusters)
    kmeansModel.fit(Xclustering)
    Ypredit = kmeansModel.predict(Xclustering)
    plt.figure()
    titulo ='El numero de clusterst es',str(n_clusters)
    plt.title(titulo)
    plt.xlabel('Caracteristica 1')
    plt.ylabel('Caracteristica 2')
    plt.scatter(Xclustering[:,0],Xclustering[:,1], c=Ypredit)    
```

```python
for n in range(2,4):
    implentationKmeans(Xclustering,n)
```
